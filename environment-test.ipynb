## 2. 에러 여부를 확인
위의 셀을 실행할 때 오류가 나타나지 않았다면 다음 단계를 진행할 수 있습니다.  

오류가 나타나면 아래 작업들을 올바르게 수행했는지 확인해야 합니다.
1. 필요 라이브러리 설치 여부 (로딩이 안되는 패키지의 경우)

2. 가상 환경 활성화 여부 
   - 패키지를 설치했는데도 import가 안 된다면, 가상환경이 활성화되지 않은 상태에서(시스템 Python에) 설치했거나 다른 가상환경에 설치했을 가능성이 큽니다.

   - 현재 사용 중인 파이썬 경로 확인: which python3 (macOS/Linux), where python (Windows)
   - 현재 pip가 가리키는 위치 확인: python3 -m pip -V
   - 우측 상단의 커널이 가상환경을 사용중인지 확인: .venv(Python 3.11.9)
